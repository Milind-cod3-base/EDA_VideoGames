# EDA: Exploratory Data Analysis

```python
#Using Matplotlib library to visualise

vg_plot = vgame_df[0:25]
vg_plot
```

```python
x = vg_plot['Rank']
y = vg_plot['Year']

plt.figure(figsize=(25,8), dpi=80)
plt.plot(x,y, label = 'Year', color = 'green')
plt.xlabel('Rank')
plt.ylabel('Year')
plt.title('Global Sales by Rank For 25 Rows')
plt.legend()
plt.show()
```
